```python
d = {
    'item name' : [],
    'Price before discount' : [],
    'Price after discount' : [],
    'ratings count' : []
}
```

```python
for i in links:
    try:    
        name = i.find('p').text
        bf_dis = i.find('span',attrs = {'class' : 'lfloat product-desc-price strike'}).text
        af_dis = i.find('span',attrs = {'class' : 'lfloat product-price'}).text
        rat = i.find('p',attrs = {'class' : 'product-rating-count'}).text
        d['item name'].append(name)
        d['Price before discount'].append(bf_dis)
        d['Price after discount'].append(af_dis)
        d['ratings count'].append(rat)
    except:
        pass
```
